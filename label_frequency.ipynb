```python
import os
import matplotlib.pyplot as plt

def visualize_labels(folder):
    labels_counter = [0,0]
    # zero_labelled = set()
    for file_name in os.listdir(folder):
        f = open(os.path.join(folder,file_name),"r")
        for line in f:
            label = line[0]
            if int(label)==0:
                # zero_labelled.add(file_name)
                pass
            labels_counter[int(label)]+=1
            
    plt.bar([0,1], labels_counter)
    plt.show()


visualize_labels(input("Enter path of labels folder: "))
```
